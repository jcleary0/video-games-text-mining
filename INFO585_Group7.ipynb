```python
input_path = "/content/drive/MyDrive/INFO585_Group7/Video_Games.jsonl.gz"
output_path = "/content/drive/MyDrive/INFO585_Group7/Video_Games.jsonl.gz"

chunks = pd.read_json(input_path, lines=True, compression="gzip", chunksize=100000)

sample_frac = 0.10
df_list = []
for chunk in chunks:
    df_list.append(chunk.sample(frac=sample_frac, random_state=42))

df_sample = pd.concat(df_list)

df_sample.to_json(output_path, orient="records", lines=True, compression="gzip")

df_sample.head()
```
